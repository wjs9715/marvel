```python
#In most social networks, two nodes that are linked to a third one have a higher probability 
#to be linked between them: two acquaintances of a given person probably know each other. 
#This effect is measured using the clustering coefficient, that is defined as follows. 
#Given a node v in a network, let kv be its degree, i.e., the number of neighbors of v, 
#and let Nv be the number of links between these kv neighbors of v. 
#If all these nodes were linked to each other, 
#then Nv would be equal to the number of unordered pairs of nodes belonging to this set of kv neighbors, 
#i.e., to kv(kv − 1)/2. The clustering coefficient Cv of node v rates the difference 
#between the actual value Nv and this greatest value by taking their quotient
#Cv= 2Nv / kv(kv −1)
#Thus, this coefficient Cv measures the fraction of neighbors of node v that are linked. 
#Notice that 0 ≤ Cv ≤ 1. The clustering coefficient C of a network is then defined 
#as the mean value of the clustering coefficients of all its nodes. 
#It represents the probability that two neighbors of an arbitrary node are linked.

#So, linkage function uses the eculidian distance to calculate the distance between two nodes
#then cluster the nodes by d(u,v)=max(dist(u[i],v[j]))(this is complete method)
#If we use the euclidian distance to calculate the distance for our graph->feature vector,
#it just turns the graph into meaningless number.

#So, what I thought we would do is get the clustering coeff of each node (from the og study),
#and use the mean probability of two node as our distance

#In the og study, coeff is definded as Cv= 2Nv / kv(kv −1)
#kv is the number of neighbours
#for example, character number 1 has only one edge 6487.
#Then, his neighbours are every characters that appeared in comic[6487]

#Nv is the link between the neighbours
#For example, character 1's neighbours are nei_stack [1998, 6458, 6462, 6463, 6470]
#and each neighbours edges are
#1998 [0   84   85   86   87   88   89   90 9750 9751 9752 9753 9754 9755 9756 9757 9758 9759]
#6458 [0]
#6462 [0    84    85    86    87    88    89    90  8180  9750  9751  9752 9753  9754  9755  9756  9757  9758  9759 10297 12940 12941]
#6463 [0    84    85    86    87    88    89    90  9750  9751  9752  9753 9754  9755  9756  9757  9758  9759 10297 12940]
#6470 [0    84    85    86    87    88    89    90  9750  9751  9752  9753 9754  9755  9756  9757  9758  9759 12940]

#links between them are
#0 not counted because it was counted as neighbour
#node 1998 has link to 6462
#node 1998 has link to 6463
#node 1998 has link to 6470
#node 6462 has link to 6463
#node 6462 has link to 6470
#node 6463 has link to 6470
#So, if every neighbours are connected, Nv equals the number of unordered pairs of neighbours


import math
    
def coeff_custom_array_test(v_array):
    node=v_array
    
    #index of node in binary feature array
    node_idx=node[0]
    #print(node_idx)

    #slice index leaves binary feature
    node=v_array[1:]
    #print(node)
    
    #To find neighbors of node, get columns of node
    #i.e) character number 1 has one edge 6487
    #every nonzero in column index 6487 is neighbour of 1    
    edge_of_node=np.nonzero(node)[0]
    #print("edge of node",edge_of_node)
    
    #check each column
    not_nei=0
    
    nei_stack=[]
    for idx in edge_of_node:
            temp_column=binary_feature[:,[idx]]
            #This is list of vertex idx that is neighbor with the tested character
            temp_column_non_zero=np.nonzero(temp_column)[0]
            #print("nonzero",temp_column_non_zero)
            node_self=np.where(temp_column_non_zero==node_idx)[0]
            #print("node_idx where",node_self)
            temp_column_non_zero=np.delete(temp_column_non_zero,node_self)
            #print("neighbour in comic column",idx,"=",temp_column_non_zero)
            
            for nei in temp_column_non_zero:
                if nei not in nei_stack:
                    nei_stack.append(nei)
                    
            #print("nei stack",nei_stack)
            
    
    #get neighbour count
    nei_count=len(nei_stack)
    #print("nei_count",nei_count)
    #print("nei_stack",nei_stack)

    link=0
    #add all    
    sum_holder=np.zeros(len(comic_book_buffer))
    for nei in nei_stack:
        #print(nei)
        sum_holder+=binary_feature[nei]
        #print(sum_holder)
    
    sum_holder[node_idx]=0 
    #print("sum initial",sum_holder)
    
    for x in range(0,nei_count-1):
        #print("ver",nei_stack[x])
        #Vertex being tested
        ver_tested=np.copy(binary_feature[nei_stack[x]])
        ver_tested[node_idx]=0
        #Edge list of vertex being tested
        test_where=np.nonzero(ver_tested)[0]
        #print(test_where)
        
        #Subtract
        sum_holder-=ver_tested
        
        if len(test_where)>0:
            sum_where=np.nonzero(sum_holder)[0]
            #print("sum",sum_where)
            temp=np.array(sum_holder[sum_where])

            #print("temp",temp)
            max_link=np.amax(temp)
            #print("max",max_link)
            link+=max_link
    
    if nei_count<=1:
        coeff=0
    else:
        coeff=(2*link)/(nei_count*(nei_count-1))
        
    #print("link",link)
       
    return coeff
    
def dis_custom(p1,p2):
    coeff=(p1+p2)/2
    return coeff

```

```python
from scipy import stats

#BRUTE
#Find number of emelemets in truth that is also in test
#Return count and perc
def matching_score(truth,test):
    
    m_truth=len(truth)
    m_test=len(test)
    m_diff=m_test
    
    count=0
    for item in truth:
        if item in test:
            count+=1
    return count, (count/m_truth)*100, m_diff

#Get the names of character reponding to the cluster 
def cluster_character_name(largest,resultingClusters):
    
    cluster_character=[]
    count=0

    for x in range(1,largest+1):
        cluster_character.append([])
        where_idx=np.where(resultingClusters==x)[0]
        for y in range(0,len(where_idx)):
            cluster_character[x-1].append(character_buffer[where_idx[y]])
    
    return cluster_character
```

```python
#BRUTE
#Iterate from t_begin to t_end
#Get score and return max of each t
def cluster_score_t_iterator(linkage_cluster,t_begin,t_end,t_step,t_criterion,truth):
    
    max_scores=[]
    count=0
    
    for t_var in np.arange(t_begin,t_end,t_step):
        #t_var*=t_step
        
        max_scores.append([])
        flat_complete = sch.fcluster(linkage_cluster,t=t_var, criterion = t_criterion)
        largest=np.amax(flat_complete)

        #GET CHARACTER NAMES
        character_name=cluster_character_name(largest,flat_complete)

        #GET SCORE
        score_buffer=[]
        
        for cluster in character_name:
            score=matching_score(truth,cluster)
            score_buffer.append(score)
            
        #FIND MAX
        score_buffer=np.array(score_buffer)
        max_score=np.amax(score_buffer[:,[1]])
        
        #TESTING
        #If max_score>50, appends all the info returned by matching_score
        if max_score>10:
            max_scores[count].append(t_var)
            
            #score with some mo info
            test=np.where(score_buffer==max_score)
            for where_idx in test[0]:
                for score_info in score_buffer[where_idx]:
                    max_scores[count].append(score_info)
            #cluster size
            max_scores[count].append(len(character_name))
        
        else:
            max_scores[count].append(t_var)
            max_scores[count].append(max_score)
        
        count+=1
    return max_scores
```
